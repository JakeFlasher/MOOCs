# Visualizing Numerical Data

In order to visualize numerical data, we donwload "[Income Per Person](https://docs.google.com/spreadsheet/pub?key=rX3Jfop_ebuY-chuMpCgmRg&output=xlsx)" data from [Gapminder](https://www.gapminder.org/data/) which pulls information from a variety of data sources. Income per person is represented in US dollars and life expectancy in years, for the year 2012. 

First we will read the data and load it into a dataframe as:

```r
# Read xlsx file and remove empty rows 
library(readxl)
rawdata <- read_excel("./data/gdp_per_capita_ppp.xlsx")
df <- rawdata[complete.cases(rawdata),]
```

```r
dim(df)
head(df, n=5)
```

```r
# Rename the header to make things easier
years <- seq(1800, 2015, by=1)
seq_years <- as.character(years)
new_header <- c('GDP', seq_years)
names(df) <- c(new_header)
```

```r
head(df, n=5)
```

```r
library(ggplot2)
plot(df, aes=(x='1812')) + geom_point()
```

```r
plot(df, mpg, main="Scatterplot Example", xlab="Car Weight ", ylab="Miles Per Gallon ", pch=19)
```

```r
plot(data=df) + geom_point()
```
