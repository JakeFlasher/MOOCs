```octave
%ploting data

t = [0:0.01:0.98];
y1 = sin(2*pi*4*t);
y2 = cos(2*pi*4*t);
plot(t,y1);
hold on; %for two or more plots 
plot(t,y2, 'r');
xlabel('time')
ylabel('value')
legend('sin', 'cos')
title('plot of sin and cos')
print -dpng 'myPlot.png'
```

```octave
figure(1); plot(t, y1);
figure(2); plot(t, y2);
```

```octave
subplot(1,2,1); % grid (1x2) access the plot 1
plot(t, y1);
subplot(1,2,2); % grid (1x2) access the plot 2
plot(t, y2);
axis([0.5 1 -1 1]); %change range of the axis [min(x),max(x),min(y),max(y)]
```
