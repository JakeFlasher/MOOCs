```python
# PROBLEM 5  (10 points possible)
# Write a Python function that creates and returns a list of prime numbers between 2 and N, inclusive, sorted in 
# increasing order. A prime number is a number that is divisible only by 1 and itself. This function takes in an 
# integer and returns a list of integers.

def primesList(N):
    '''
    N: an integer
    Returns a list of prime numbers
    '''
    primes = []
    for nb in range(2,N+1):
        valid = True
        for dv in range(2,nb):
            if nb % dv == 0 and dv != 1 and dv != nb:
                valid = False
        if valid:
            primes.append(nb)
    return primes

print primesList(11)
```

```python
# PROBLEM 6  (10 points possible)
# Write a recursive Python function, given a non-negative integer N, to count and return the number of occurrences 
# of the digit 7 in N.
#
# For example:
# count7(717) -> 2
# count7(1237123) -> 1
# count7(8989) -> 0
# Hint: Mod (%) by 10 gives you the rightmost digit (126 % 10 is 6), while doing integer division by 10 removes the 
# rightmost digit (126 / 10 is 12).
#
# This function has to be recursive; you may not use loops! This function takes in one integer and returns one integer.

def count7(N):
    '''
    N: a non-negative integer
    '''
    if len(str(N)) == 1:
        if N % 10 == 7:
            return 1
        else:
            return 0
    else:
        if N % 10 == 7:
            return 1 + count7(N/10)
        else:
            return count7(N/10)
        
print count7(12323)
print count7(717)
print count7(1237123)
print count7(7718294771)
        
```

```python
# PROBLEM 7  (20 points possible)
# Write a Python function that returns a list of keys in aDict that map to integer values that are unique 
# (i.e. values appear exactly once in aDict). The list of keys you return should be sorted in increasing 
# order. (If aDict does not contain any unique values, you should return an empty list.)
#
# This function takes in a dictionary and returns a list.

def uniqueValues(aDict):
    '''
    aDict: a dictionary
    '''
    unique = []
    for key in aDict:
        v1 = aDict[key]
        sumv = 0
        for v2 in aDict.values():
            if v1 == v2:
                sumv += 1
        if sumv == 1:
            unique.append(key)
    return sorted(unique)

print uniqueValues({1:10, 2:20, 3:30, 4:10, 5:40})
print uniqueValues({1:10, 2:20, 3:30, 4:40})
print uniqueValues({1:10, 2:20, 3:30, 4:10, 5:30, 6:20})
```
