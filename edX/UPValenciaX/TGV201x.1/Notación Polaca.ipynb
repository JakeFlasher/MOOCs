```python
# generate the expression based on a "notación polaca"
# from the input. The function performs the inverse of
# notacion_polaca() function.
# The input is an expression in "notación polaca" or
# notación polaca inversa such as:
# 
# 3 x 4 * - 5 y 2 * x + ^ /
# 
# and the output is a contructed expression in the form:
# 
# (((x+2)*y)^5)/((4*x)-3)

def generate_expression(notacion_polaca, inverse=False):
    """
    input: string containing the "notación polaca"
    output: string containing the expression
    """
    symbols = ['*', '-', '+', '/', '^']
    notacion = list(notacion_polaca)
    while len(notacion) > 1:
        index = 0
        tmp_notacion = []
        for el in notacion:
            if inverse:
                if el not in symbols \
                   and notacion[index+1] not in symbols \
                   and notacion[index+2] in symbols:
                    tmp_notacion.append('(%s %s %s)' % (notacion[index+1], notacion[index+2], el))
                    tmp_notacion.extend(notacion[index+3:])
                    break
                else:
                    tmp_notacion.append(el)
                    index += 1
            else:
                if el in symbols \
                   and notacion[index+1] not in symbols \
                   and notacion[index+2] not in symbols:
                    tmp_notacion.append('(%s %s %s)' % (notacion[index+1], el, notacion[index+2]))
                    tmp_notacion.extend(notacion[index+3:])
                    break
                else:
                    tmp_notacion.append(el)
                    index += 1
        notacion = tmp_notacion[:]
        tmp_notacion = []
    return ''.join(notacion)


#notacion_polaca = '/^+x*2y5-*4x3'
notacion_polaca = '/*x2^5+2*x3'
print generate_expression(notacion_polaca, inverse=False)

#notacion_polaca_inversa = '3x4*-5y2*x+^/'
notacion_polaca_inversa = '3x*2+5^2x*/'
print generate_expression(notacion_polaca_inversa, inverse=True)
```
