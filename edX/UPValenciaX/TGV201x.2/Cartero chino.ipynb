```python
import networkx as nx

G = nx.Graph()
nodes = ['v1', 'v2', 'v3', 'v4', 'v5', 'v6', 'v7', 'v8', 'v9', 'v10', 'v11', 'v12']
G.add_nodes_from(nodes)

G.add_edge('v1','v2', weight=1)
G.add_edge('v1','v4', weight=2)
G.add_edge('v2','v5', weight=2)
G.add_edge('v3','v4', weight=3)
G.add_edge('v3','v7', weight=4)
G.add_edge('v4','v5', weight=3)
G.add_edge('v4','v8', weight=4)
G.add_edge('v5','v6', weight=3)
G.add_edge('v5','v9', weight=4)
G.add_edge('v6','v10', weight=4)
G.add_edge('v7','v8', weight=3)
G.add_edge('v8','v9', weight=3)
G.add_edge('v8','v11', weight=5)
G.add_edge('v9','v10', weight=3)
G.add_edge('v9','v12', weight=5)
G.add_edge('v11','v12', weight=1)

nx.draw(G)
```
