```python
# Algorithm to color a graph where two nodes that are connected through an edge 
# cannot share the same color

import networkx as nx

def init(nodes, edges):
    G = nx.Graph()
    G.add_nodes_from(nodes, color=0.0)
    for n1, n2 in edges:
        G.add_edge(n1, n2)
    return G

def print_info(G):
    print 'Nodes:', G.nodes()
    print  [(k, round(G.node[k]['color'], 3)) for k in G.nodes()]
    print 

def graph_color(G, info=True):
    nodes = sorted(G.nodes())
    palette = [random.random() for _ in nodes]
    G.node[nodes[0]]['color'] = palette[0]
    if info: print_info(G)
   
    for node in sorted(G.nodes()):
        rejected = []
        for neighbor in G.neighbors(node):
            if G.node[neighbor]['color']:
                rejected.append(G.node[neighbor]['color'])
        for col in palette:
            if col not in rejected:
                G.node[node]['color'] = col
                break
        if info: print_info(G)
            
    colors = [G.node[node]['color'] for node in G.nodes()]
    return colors

def draw_graph(G, colors=False, info=False):
    if colors:
        colors = graph_color(G, info=info)
        nx.draw(G, node_color=colors)
    else:
        nx.draw(G)
```

```python
# Ejemplo 4: \chi(Cn) = \left\{\begin{matrix}
#                                 2 & \text{si n és par}\\ 
#                                 3 & \text{si n és impar}
#                              \end{matrix}\right.
# (grafo circular)

nodes = ['A', 'B', 'C', 'D', 'E']
edges = [('A','B'), ('B','C'), ('C','D'), ('D','E'), ('E', 'A')]
G = init(nodes, edges)
plt.figure(1, figsize=(3, 3), dpi=80, facecolor='w', edgecolor='k')
draw_graph(G, colors=False, info=False)
plt.figure(2, figsize=(3, 3), dpi=80, facecolor='w', edgecolor='k')
draw_graph(G, colors=True, info=False)
```
