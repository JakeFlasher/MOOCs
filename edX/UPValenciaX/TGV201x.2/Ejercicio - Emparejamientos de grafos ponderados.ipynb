```python
import networkx as nx
```

```python
# EMPAREJAMIENTOS EN GRAFOS BIPARTIDOS PONDERADOS (MODELIZACIÓN) (1 point possible)
# Se necesita contratar a 4 programadores para programar en 4 lenguajes distintos: 
# C, Java, Lisp y Python. A cada uno de los 4 candidatos, Andrés, Eduardo, Emilio y Juan Carlos, 
# les hemos hecho unas pruebas para saber su nivel de competencia en cada lenguaje, valorándolos de 
# 0 (conocimiento nulo) a 10 (nivel experto). Los resultados se muestran en la siguiente tabla:
#
# 
#              C  Java  Lisp  Python
# Andrés       9   7     6      8
# Eduardo      4   7     3      6
# Emilio       8   4     5      5
# Juan Carlos  2   5     9      7
#
# Nos proponen como solución óptima la siguiente:
# Andrés - C, Eduardo - Java, Emilio - Lisp y Juan Carlos - Python.
# No obstante, queremos asegurarnos de que no hay una mejor. Para ello consideramos un grafo bipartido 
# G=((X,Y),E) donde X={programadores}, Y={lenguajes} y E={(x,y)/ x en X, y enY, x sabe programar en y} 
# y buscamos un emparejamiento máximo de máximo peso mediante uno de los algoritmos vistos.
#
# ¿Es la solución propuesta óptima? 

G = nx.Graph()
nodes = ['Andres', 'Eduardo', 'Emilio', 'Juan', 'C', 'Java', 'Lisp', 'Python']
colors = ['r','r','r','r','b','b','b','b']
G.add_nodes_from(nodes)

G.add_edge('Andres','C', weight=9)
G.add_edge('Andres','Java', weight=7)
G.add_edge('Andres','Lisp', weight=6)
G.add_edge('Andres','Python', weight=8)
G.add_edge('Eduardo','C', weight=4)
G.add_edge('Eduardo','Java', weight=7)
G.add_edge('Eduardo','Lisp', weight=3)
G.add_edge('Eduardo','Python', weight=6)
G.add_edge('Emilio','C', weight=8)
G.add_edge('Emilio','Java', weight=4)
G.add_edge('Emilio','Lisp', weight=5)
G.add_edge('Emilio','Python', weight=5)
G.add_edge('Juan','C', weight=2)
G.add_edge('Juan','Java', weight=5)
G.add_edge('Juan','Lisp', weight=9)
G.add_edge('Juan','Python', weight=7)

nx.draw_circular(G, node_color=colors)
```

```python
maxG = nx.max_weight_matching(G, maxcardinality=True)

nG = nx.Graph()
nG.add_nodes_from(nodes)
nG.add_edges_from(list(maxG.items()))
nx.draw(nG, node_color=colors)
```

```python
# EMPAREJAMIENTOS EN GRAFOS PONDERADOS (MODELIZACIÓN)  (1 point possible)
# Las 8 alumnas de una escuela de danza van a representar una obra para la que es necesario en determinados 
# momentos realizar figuras por parejas. El entrenador construye la siguiente tabla en la que refleja la compatibilidad 
# entre las chicas a la hora de realizar las figuras de baile. Las casillas en blanco reflejan que la estatura u otras 
# características hacen no recomendable la pareja en cuestión. 
#
#        Sonia  Ana  Paz  Olga  Luisa  Mila  Rosa  Elsa
# Sonia    -     5    6    -      9     -     -     2
# Ana      5     -    -    7      -     8     -     -
# Paz      6     -    -    -      6     -     8     7
# Olga     -     7    -    -      -     3     -     -
# Luisa    9     -    6    -      -     -     5     10
# Mila     -     8    -    3      -     -     6     -
# Rosa     -     -    8    -      5     6     -     2
# Elsa     2     -    7    -      10    -     2     -
#
# La suma de pesos del emparejamiento {(Sonia, Luisa), (Ana, Olga), (Paz, Elsa), (Mila, Rosa)} es 29. 
# ¿Es el emparejamiento de mayor peso? 

G = nx.Graph()
nodes = ['Sonia', 'Ana', 'Paz', 'Olga', 'Luisa', 'Mila', 'Rosa', 'Elsa']
colors = ['y','y','y','y','y','y','y','y']
G.add_nodes_from(nodes)

G.add_edge('Sonia','Ana', weight=5)
G.add_edge('Sonia','Paz', weight=6)
G.add_edge('Sonia','Luisa', weight=9)
G.add_edge('Sonia','Elsa', weight=2)
G.add_edge('Ana','Sonia', weight=5)
G.add_edge('Ana','Olga', weight=7)
G.add_edge('Ana','Mila', weight=8)
G.add_edge('Paz','Sonia', weight=6)
G.add_edge('Paz','Luisa', weight=6)
G.add_edge('Paz','Rosa', weight=8)
G.add_edge('Paz','Elsa', weight=7)
G.add_edge('Olga','Ana', weight=7)
G.add_edge('Olga','Mila', weight=3)
G.add_edge('Luisa','Sonia', weight=9)
G.add_edge('Luisa','Paz', weight=6)
G.add_edge('Luisa','Rosa', weight=5)
G.add_edge('Luisa','Elsa', weight=10)
G.add_edge('Mila','Ana', weight=8)
G.add_edge('Mila','Olga', weight=3)
G.add_edge('Mila','Rosa', weight=6)
G.add_edge('Rosa','Paz', weight=8)
G.add_edge('Rosa','Luisa', weight=5)
G.add_edge('Rosa','Mila', weight=6)
G.add_edge('Rosa','Elsa', weight=2)
G.add_edge('Elsa','Sonia', weight=2)
G.add_edge('Elsa','Paz', weight=7)
G.add_edge('Elsa','Luisa', weight=10)
G.add_edge('Elsa','Rosa', weight=2)


nx.draw_circular(G, node_color=colors)
```

```python
maxG = nx.max_weight_matching(G, maxcardinality=True)

nG = nx.Graph()
nG.add_nodes_from(nodes)
nG.add_edges_from(list(maxG.items()))
nx.draw(nG, node_color=colors)
```
