```python
# Cinco de nuestros amigos, María, Sergio, Álvaro, Sonia y Luisa, se van a presentar a un concurso internacional 
# de preguntas – respuestas sobre 7 materias: Matemáticas, Física, Historia, Química, Geografía, Biología y Economía. 
# En cada turno se plantearán sendas preguntas de estos 7 temas. Gana el turno el equipo que más rápidamente responda 
# correctamente a 5 de ellas. La estrategia por tanto es que cada uno de nuestros amigos elija un tema y se lo prepare 
# exhaustivamente. Para decidir cuál debe prepararse cada uno construyen la siguiente tabla en la que han reflejado su 
# grado de conocimiento. ¿Puedes ayudarles a decidir qué tema debe estudiar cada uno y cuáles quedarán sin estudiar a 
# fin de tener más probabilidades de ganar?
#
#        Matemáticas  Física  Historia  Química  Literatura  Biología  Filosofía
# María      -          5        -         2          -         2          -         
# Sergio     2          -        2         2          -         2          -
# Álvaro     2          -        7         -          -         -          -
# Elena      -          5        -         -          2         -          3
# Luisa      4          -        1         -          1         1          3
#
# Indica cuál de las siguientes respuestas no es correcta.
#
# 1. Una de las mejores elecciones sería {(María, Física), (Sergio, Biología), (Álvaro, Historia), (Elena, FIlosofía), 
#    (Luisa, Matemáticas)}.  
# 2. Para resolver el problema se debe aplicar el algoritmo de Kuhn Munkres a la matriz dada (NOTA: Nos referimos a la 
#    tabla dada considerada como matriz).  
# 3. El problema se puede resolver con el programa SWGraphs ejecutando el algoritmo de Edmonds para máximo peso.  
# 4. Una buena opción sería {(María, Física), (Sergio, Química), (Álvaro, Historia), (Elena, FIlosofía), (Luisa, Matemáticas)}.

import networkx as nx

G = nx.Graph()
nodes = ['Maria', 'Sergio', 'Alvaro', 'Elena', 'Luisa', 'Matematicas', 'Fisica', 'Historia', 'Quimica', 'Literatura', 'Biologia', 'Filosofia']
colors = ['b','r','b','b','r','b','r','b','r','b','b','r']
G.add_nodes_from(nodes)

G.add_edge('Maria','Fisica', weight=5)
G.add_edge('Maria','Quimica', weight=2)
G.add_edge('Maria','Biologia', weight=2)
G.add_edge('Sergio','Matematicas', weight=2)
G.add_edge('Sergio','Historia', weight=2)
G.add_edge('Sergio','Quimica', weight=2)
G.add_edge('Sergio','Biologia', weight=2)
G.add_edge('Alvaro','Matematicas', weight=2)
G.add_edge('Alvaro','Historia', weight=7)
G.add_edge('Elena','Fisica', weight=5)
G.add_edge('Elena','Literatura', weight=2)
G.add_edge('Elena','Filosofia', weight=3)
G.add_edge('Luisa','Matematicas', weight=4)
G.add_edge('Luisa','Historia', weight=1)
G.add_edge('Luisa','Literatura', weight=1)
G.add_edge('Luisa','Biologia', weight=1)
G.add_edge('Luisa','Filosofia', weight=3)

nx.draw_circular(G, node_color=colors)
```

```python
maxG = nx.max_weight_matching(G, maxcardinality=True)

nG = nx.Graph()
nG.add_nodes_from(nodes)
nG.add_edges_from(list(maxG.items()))
nx.draw(nG, node_color=colors)
```
