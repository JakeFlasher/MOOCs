```python
# La empresa de productos textiles ImprimeTEX quiere planificar las jornadas de trabajo del siguiente año. 
# Cada día de la semana, de lunes a viernes, debe dar una determinada coloración a distintas prendas de vestir. 
# La empresa utiliza 5 colores y para adecuar la maquinaria de un color a otro existen unos costes de limpieza 
# dependiendo del color de que se trata. En cada jornada de trabajo, utilizan únicamente un color y al final 
# de ella,cuando han acabado con todas las piezas de ropa es necesario limpiar la maquinaria para dejarla 
# preparada para empezar a trabajar con otro color al día siguiente. Los fines de semana también debe dejarse 
# limpia, ya que en otro caso, al secarse la pintura podría estropear la maquinaria.¿Cuál sería el orden más 
# conveniente que la empresa debe seguir? Los costes de limpieza están reflejados en la siguiente tabla:
# Coste de limpiar del color (ver fila) al color (ver columna)
#
#          Amarillo  Azul  Negro  Rojo  Verde
# Amarillo	0	   9	  5	 17	 5
# Azul	    9	   0	 14	  8	11
# Negro	   5	  14	  0	 12	10
# Rojo	   17	   8	 12	  0	12
# Verde	   5	  11	 10	 12	 0
#
# Para resolver este problema ayúdate de la soluciones de peso bajo al problema del viajante.

import networkx as nx

def init():
    G = nx.Graph()
    nodes = ['Amarillo', 'Azul', 'Negro', 'Rojo', 'Verde']
    G.add_nodes_from(nodes)
    
    G.add_edge('Amarillo','Azul', weight=9)
    G.add_edge('Amarillo','Negro', weight=5)
    G.add_edge('Amarillo','Rojo', weight=17)
    G.add_edge('Amarillo','Verde', weight=5)
    G.add_edge('Azul','Negro', weight=14)
    G.add_edge('Azul','Rojo', weight=8)
    G.add_edge('Azul','Verde', weight=11)
    G.add_edge('Negro','Rojo', weight=12)
    G.add_edge('Negro','Verde', weight=10)
    G.add_edge('Rojo','Verde', weight=12)
    return G

G = init()
nx.draw_circular(G)
```

```python
def print_info(step, nodes, start, nmin, wmin, C):
    print 'step: ', step
    print 'nodes: ',nodes
    print 'start:', start
    print 'nmin: ', nmin
    print 'wmin: ', wmin
    print 'C: ', C
    print '---------------'
    
    
def viajante(G, start=None, full=True, info=False):
    nodes = G.nodes()
    step = 1
    if start:
        current = start
    else:
        current = nodes[0]
    C = [current]
    wmax_tmp = max([e['weight'] for _, _, e in G.edges_iter(data=True)])
    if info:
        print_info(step, nodes, current, '', wmax_tmp, C)
    while len(nodes) > 1:
        wmin = wmax_tmp
        nmin = ''
        for neighbor in G.neighbors(current):
            weight = G[current][neighbor]['weight']
            if weight <= wmin:
                nmin = neighbor
                wmin = weight
        C.insert(0, nmin)
        for neighbor in G.neighbors(current):
            G.remove_edge(current, neighbor)
        nodes.remove(current)
        step += 1
        if info:
            print_info(step, nodes, current, nmin, wmin, C)
        current = nmin
    if full:
        C.insert(0, start)
    return C

def path_weight(G, path, info=False):
    weight = 0
    for i, el1 in enumerate(path):
        if i < len(path)-1:
            el2 = path[i+1]
            weight += G[el1][el2]['weight']
            if info:
                print '(%s, %s): %d' % (el1, el2, G[el1][el2]['weight'])
    return weight
            
G = init()
path = viajante(G.copy(), 'Amarillo', full=True, info=False)
print 'Path: ', path
weight = path_weight(G, path, info=True)
print 'Weight: ', weight
```
