```python
import networkx as nx

G = nx.Graph()

nodes = ['E', 'C', 'A', 'S', 'T', 'D', 'P', 'G']
edges = [('E','T'), ('E','P'), ('E','G'), ('C','T'), ('C','P'), ('A','D'), ('S','D')]
G.add_nodes_from(nodes)
G.add_edges_from(edges)

nx.draw(G)
```

```python
maxG = nx.max_weight_matching(G, maxcardinality=False)

nG = nx.Graph()
nG.add_nodes_from(nodes)
nG.add_edges_from(list(maxG.items()))
nx.draw(nG)
```
