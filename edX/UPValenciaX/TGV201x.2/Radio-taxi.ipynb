```python
import networkx as nx
```

```python
# PARTE 1:
#
# Taxis: T1, T2, T3, T4
# Clientes: C1, C2, C3, C4

G = nx.Graph()
nodes = ['T1', 'T2', 'T3', 'T4', 'C1', 'C2', 'C3', 'C4']
colors = ['r','r','r','r','b','b','b','b']
G.add_nodes_from(nodes)

# weights are negative because it is looking for the minimum weight
G.add_edge('T1','C1', weight=-6)
G.add_edge('T1','C2', weight=-5)
G.add_edge('T1','C3', weight=-7)
G.add_edge('T1','C4', weight=-6)
G.add_edge('T2','C1', weight=-8)
G.add_edge('T2','C2', weight=-5)
G.add_edge('T2','C3', weight=-4)
G.add_edge('T2','C4', weight=-12)
G.add_edge('T3','C1', weight=-7)
G.add_edge('T3','C2', weight=-7)
G.add_edge('T3','C3', weight=-5)
G.add_edge('T3','C4', weight=-11)
G.add_edge('T4','C1', weight=-7)
G.add_edge('T4','C2', weight=-8)
G.add_edge('T4','C3', weight=-6)
G.add_edge('T4','C4', weight=-10)

nx.draw_circular(G, node_color=colors)
```

```python
maxG = nx.max_weight_matching(G, maxcardinality=True)

nG = nx.Graph()
nG.add_nodes_from(nodes)
nG.add_edges_from(list(maxG.items()))
nx.draw(nG, node_color=colors)
```

```python
# PARTE 2
#
# Si considera que los valores mayores que 6 no se deben tener en cuenta por considerarse excesivos

G = nx.Graph()
nodes = ['T1', 'T2', 'T3', 'T4', 'C1', 'C2', 'C3', 'C4']
colors = ['r','r','r','r','b','b','b','b']
G.add_nodes_from(nodes)

# weights are negative because it is looking for the minimum weight
G.add_edge('T1','C1', weight=-6)
G.add_edge('T1','C2', weight=-5)
G.add_edge('T1','C4', weight=-6)
G.add_edge('T2','C2', weight=-5)
G.add_edge('T2','C3', weight=-4)
G.add_edge('T3','C3', weight=-5)
G.add_edge('T4','C3', weight=-6)

nx.draw_circular(G, node_color=colors)
```
